## Different between in age groups

```python
# Assuming you have a 'ratings' DataFrame with 'user_id', 'rating', and a 'users' DataFrame with 'user_id' and 'age'

# Merge 'ratings' and 'users' on 'user_id'
merged_data = pd.merge(ratings, users[['user_id', 'age']], on='user_id', how='left')

# Define age groups
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+']

# Create 'Age Group' column
merged_data['Age Group'] = pd.cut(merged_data['age'], bins=age_bins, labels=age_labels, right=False)

# Calculate average ratings by age group
average_ratings_by_age = merged_data.groupby('Age Group')['rating'].mean()

# Visualize average ratings by age group
plt.figure(figsize=(10, 6))
sns.barplot(x=average_ratings_by_age.index, y=average_ratings_by_age.values, palette='viridis')
plt.title('Average Ratings by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Rating')
plt.show()
```

### Average Ratings by Age Group:

Objective: Understand how users' age correlates with the average ratings they give to books.

Insights:
Identify if there are variations in average ratings across different age groups.
Explore whether users in certain age groups tend to give higher or lower ratings on average.

Use Case:
This information can be valuable for tailoring book recommendations based on user age preferences. For example, you might adjust recommendations to favor books that are popular among users in specific age brackets.

```python
# Calculate the number of ratings by age group
ratings_by_age = merged_data.groupby('age')['rating'].count()

# Visualize the distribution of ratings by age group
plt.figure(figsize=(10, 6))
sns.barplot(x=ratings_by_age.index, y=ratings_by_age.values, palette='magma')
plt.title('Number of Ratings by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Ratings')
plt.show()
```

### Number of Ratings by Age Group:

Objective: Explore the distribution of user engagement with the recommendation system across different age groups.

Insights:
Identify which age groups are more active in providing ratings.
Assess the overall level of user engagement in the recommendation system for each age group.

Use Case:
This information can help you understand whether certain age groups are more active in the system, which might influence how you prioritize and interpret user feedback. It can also guide strategies for increasing engagement among less active age groups.

# Model Building

# Training Model

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
# Now import our clustering algoritm which is Nearest Neighbors this is an unsupervised ml algo
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6 )
```

# Testing model

```python
def recommend_book(book_name):
     # Find the index of the book in the pivot table
    book_id = np.where(book_pivot.index == book_name)[0][0]
    # Find the nearest neighbors (books) based on the similarity of ratings
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    # Iterate over the suggestions and print the book names
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = 2766
recommend_book(book_name)
```

### Explanation:

1. Finding the Index of the Searched Book:
2. Finding Nearest Neighbors:
3. Iterating Over Suggestions and Printing:

#### Finding the Book's Position:

book_id = np.where(book_pivot.index == book_name)[0][0]: This line finds the position (index) of the searched book in the list of all books.

#### Finding Similar Books:

distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6): This line finds other books that are similar to the searched book based on user ratings.

#### Printing Searched Book:

print(f"You searched for '{book_name}'\n"): This line prints the name of the book you searched for.

#### Printing Suggested Books:

print("Suggested books are: \n"): This line indicates that the following list will contain suggested books.
The loop then iterates through the suggested books and prints each one, excluding the searched book.
