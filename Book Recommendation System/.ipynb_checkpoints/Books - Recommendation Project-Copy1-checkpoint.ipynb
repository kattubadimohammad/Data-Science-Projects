# Feature Engineering

```python
# 1. User-Item Interaction Features
# Number of ratings per user and per book
user_ratings_count = ratings.groupby('user_id').size().reset_index(name='user_ratings_count')
book_ratings_count = ratings.groupby('ISBN').size().reset_index(name='book_ratings_count')
```

```python
# Average rating per user and per book
user_avg_rating = ratings.groupby('user_id')['rating'].mean().reset_index(name='user_avg_rating')
book_avg_rating = ratings.groupby('ISBN')['rating'].mean().reset_index(name='book_avg_rating')
```

```python
# Merge the new features into the users and books dataframes
users = pd.merge(users, user_ratings_count, on='user_id', how='left')
users = pd.merge(users, user_avg_rating, on='user_id', how='left')
```

```python
books = pd.merge(books, book_ratings_count, on='ISBN', how='left')
books = pd.merge(books, book_avg_rating, on='ISBN', how='left')
```

```python
# 2. Temporal Features
# Extract the year, month, and day from the timestamp (assuming 'timestamp' is the correct column name)
if 'timestamp' in ratings.columns:
    ratings['timestamp'] = pd.to_datetime(ratings['timestamp'])
    ratings['year'] = ratings['timestamp'].dt.year
    ratings['month'] = ratings['timestamp'].dt.month
    ratings['day'] = ratings['timestamp'].dt.day
```

```python
# 3. Genre/Category Features (Assuming 'genre' is a column in the books dataframe)
# Count the number of books in each genre
genre_counts = books['title'].value_counts().reset_index(name='genre_counts')
```

```python
# Merge the genre counts into the books dataframe
books = pd.merge(books, genre_counts, left_on='title', right_on='index', how='left')
books.drop('index', axis=1, inplace=True)
```

```python
# 4. Content-Based Features (Assuming 'description' is a column in the books dataframe)
# Example: Extract the length of book descriptions
books['description_length'] = books['publisher'].apply(lambda x: len(str(x)))
```

```python
# 5. User Behavior Features
# Example: Calculate the recency of user interactions
if 'timestamp' in ratings.columns:
    latest_interaction_date = ratings['timestamp'].max()
    ratings['recency'] = (latest_interaction_date - ratings['timestamp']).dt.days
```

```python
user_recency = ratings.groupby('user_id')['rating'].min().reset_index(name='user_recency')
```

```python
# Merge the user recency into the users dataframe
users = pd.merge(users, user_recency, on='user_id', how='left')
```

```python
# 6. Collaborative Filtering Features
# Example: Calculate item similarity based on user ratings
item_similarity = ratings.groupby('ISBN')['rating'].mean().reset_index(name='item_similarity')
```

```python
# Merge the item similarity into the books dataframe
books = pd.merge(books, item_similarity, on='ISBN', how='left')
```

# Training Model

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
# Now import our clustering algoritm which is Nearest Neighbors this is an unsupervised ml algo
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6 )
```

# Testing model

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
recommend_book(book_name)
```
